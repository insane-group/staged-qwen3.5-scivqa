```python
import outlines

from transformers import AutoModelForCausalLM, AutoTokenizer
from outlines.types import CFG

from pathlib import Path
```

```python
MODEL_ID = "unsloth/Qwen3.5-0.8B"

BASE_DIR = Path.cwd()

MAX_NEW_TOKENS = 512
```

```python
model = outlines.from_transformers(
    AutoModelForCausalLM.from_pretrained(MODEL_ID, device_map="auto"),
    AutoTokenizer.from_pretrained(MODEL_ID),
)
```

```python
arithmetic_grammar = """
?start: sum

?sum: product
| sum "+" product   -> add
| sum "-" product   -> sub

?product: atom
| product "*" atom  -> mul
| product "/" atom  -> div

?atom: NUMBER           -> number
| "-" atom         -> neg
| "(" sum ")"

%import common.NUMBER
%import common.WS_INLINE

%ignore WS_INLINE
"""

output_type = CFG(arithmetic_grammar)
```

```python
result = model("Write an addition.", output_type, max_new_tokens=100)
print(result)  # '23 + 48'
```
